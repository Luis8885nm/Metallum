
## Tu Equipo

Cada equipo tiene una configuración única que genera datos diferentes pero comparables.

**Archivo de configuración**: `configs/teams/team0X.yaml`

Parámetros clave:
- `team_id`: Identificador del equipo
- `seed`: Semilla para reproducibilidad
- `population_size`: Número de usuarios
- `churn_base_rate`: Tasa base de cancelación
- Y muchos más...
